```python
# Dependencies
import pandas as pd
```

```python
# load CSV
input_file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
org_df = pd.read_csv(input_file, low_memory=False)
```

```python
# Inspect all columns
org_df.head()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_df = org_df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]].copy()
new_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
def replacing_zero(list):
    for x in range(len(list)):
        new_df[list[x]].replace({0.0:'No', 1.0:'Yes'}, inplace=True)
```

```python
required_columns = ['AttendedBootcamp', 
                    'BootcampFinish', 
                    'BootcampFullJobAfter', 
                    'BootcampLoanYesNo', 
                    'BootcampRecommend', 
                    'CodeEventBootcamp',
                    'HasChildren']
```

```python
replacing_zero(required_columns)
```

```python
new_df.head()
```

```python
# Calculate total number of respondents in survey
respondents = len(new_df)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_df = new_df.loc[new_df.AttendedBootcamp == 'Yes', :].copy()
attended_df.head()
```

```python
# Calculate average age of attendees
average_age = round(attended_df.Age.mean(),0)
average_age
```

```python
# Calculate how many people attended a bootcamp
people_who_attended = len(attended_df)
people_who_attended
```

```python
# Calculate how many attendees hold degrees
attended_df.SchoolDegree.value_counts()
```

```python
people_with_degrees = attended_df['SchoolDegree'].count()
people_with_degrees
```

```python
percentage_of_degrees = people_with_degrees / people_who_attended * 100
percentage_of_degrees
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
attended_df.Gender.value_counts()
```

```python
male = sum(attended_df['Gender'] == 'male')

female = sum(attended_df['Gender'] == 'female')

non_spec = attended_df['Gender'].count() - male - female
```

```python
# Calculate percentage of respondents who attended a bootcamp
percentage_of_attended = people_who_attended / respondents * 100
percentage_of_attended
```

```python
# Calculate percentage of respondents belonging to each gender
percentage_male = male / people_who_attended * 100
percentage_female = female / people_who_attended * 100
percentage_non_spec = non_spec / people_who_attended * 100
```

```python
# Calculate percentage of attendees with a school degree
people_with_school_degrees = sum((attended_df['SchoolDegree'] == 'high school diploma or equivalent (GED)') | 
                                (attended_df['SchoolDegree'] == 'some high school') | 
                                (attended_df['SchoolDegree'] == 'no high school (secondary school)'))

percentage_of_school_degrees = people_with_school_degrees / people_who_attended * 100
percentage_of_school_degrees
```

```python
# Calculate average post-bootcamp salary
average_post_salary = attended_df.BootcampPostSalary.mean()
average_post_salary
```

```python
# Create a new table consolodating above calculations
summary_df = pd.DataFrame({ 
    "Total Surveyed": [respondents],
    "Total Bootcamp attendees": [people_who_attended],
    "% attended bootcamp": [percentage_of_attended],
    "Avg. Age": [average_age],
    "Has a degree": [percentage_of_degrees],
    "% Male": [percentage_male],
    "% Female": [percentage_female],
    "% Non Gender Specific": [percentage_non_spec],
    "Average Post Bootcamp Salary": [average_post_salary]
})

summary_df 
```

```python
# Improve formatting before outputting spreadsheet
summary_df["% attended bootcamp"] = summary_df["% attended bootcamp"].map("{:.2f}%".format)
summary_df["Has a degree"] = summary_df["Has a degree"].map("{:.2f}%".format)
summary_df["% Male"] = summary_df["% Male"].map("{:.2f}%".format)
summary_df["% Female"] = summary_df["% Female"].map("{:.2f}%".format)
summary_df["% Non Gender Specific"] = summary_df["% Non Gender Specific"].map("{:.2f}%".format)
summary_df["Average Post Bootcamp Salary"] = summary_df["Average Post Bootcamp Salary"].map("${:,.0f}".format)

summary_df
```

```python
# Export to Excel
output_file = "Output/BootcampOutputPart1.csv"
summary_df.to_csv(output_file, index=False)
```
