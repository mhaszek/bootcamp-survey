```python
# Dependencies
import pandas as pd
```

```python
# load CSV
input_file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
org_df = pd.read_csv(input_file, low_memory=False)
org_df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_df = org_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]].copy()
new_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
def replacing_zero(list):
    for x in range(len(list)):
        new_df[list[x]].replace({0.0:'No', 1.0:'Yes'}, inplace=True)
```

```python
required_columns = ['AttendedBootcamp', 
                    'BootcampFinish', 
                    'BootcampFullJobAfter', 
                    'BootcampLoanYesNo', 
                    'BootcampRecommend', 
                    'CodeEventBootcamp',
                    'HasChildren']
```

```python
replacing_zero(required_columns)
```

```python
new_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_df = new_df.loc[new_df.AttendedBootcamp == 'Yes', :].copy()
attended_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bootcamps_attended_df = attended_df.groupby('BootcampName')['AttendedBootcamp'].count().reset_index(name='Count')

bootcamps_attended_df.head()
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
bootcamps_recommend_df = attended_df.groupby('BootcampName')['BootcampRecommend'].apply(lambda x: x[x == 'Yes'].count()).reset_index(name='Recommenders')

bootcamps_recommend_df.head()
```

```python
# Merge the two created data frames on the name of the bootcamp
summary_df = pd.merge(bootcamps_attended_df, bootcamps_recommend_df, on="BootcampName", how="outer")

summary_df.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
summary_df['% Recommend'] = summary_df['Recommenders'] / summary_df['Count'] * 100

summary_df = summary_df.sort_values(["% Recommend", 'Count'], ascending=False)

# use map() to format required columns in the DataFrame
summary_df['% Recommend'] = summary_df['% Recommend'].map("{:.2f}%".format)

summary_df.head()
```

```python
# Export to excel and remove index
output_file = "Output/BootcampOutputPart2.csv"
summary_df.to_csv(output_file, index=False)
```
